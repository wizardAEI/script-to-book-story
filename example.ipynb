```python
from PIL import Image
from sentence_transformers import SentenceTransformer, util
```

```python
image_model = SentenceTransformer('clip-VIT-B-32')
```

```python
text_model = SentenceTransformer('clip-Vit-B-32-multilingual-v1')
```

```python
img0 = Image.open('./0.png')
img1 = Image.open('./1.png')
img2 = Image.open('./2.png')
img3 = Image.open('./3.png')
img0
```

```python
text_vec = text_model.encode(['cute big yes'])
```

```python
image_vecs = image_model.encode([img0, img1, img2, img3])
```

```python
for i in range(4):
  s = util.cos_sim(text_vec, image_vecs[i])
  print(i, s)
```
